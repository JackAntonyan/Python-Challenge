```python
# Import Library
import os
import csv
# Path of the file
file = "budget_data.csv"

# Join the path
csvpath = os.path.join("budget_data.csv")

# Empty Lists
total_months = []
total_profit = []
monthly_profit_change = []

# Open and read the csv file
with open(file,newline ="") as csv_file:
    csvreader = csv.reader(csv_file, delimiter = ",")

    # Skip the header to iterate through rows
    header = next(csvreader)
      
    # Iterate through rows and append the total the months and profits to their corresponding lists
    for row in csvreader:
        total_months.append(row[0])
        total_profit.append(int(row[1]))
        
    # Iterate through the profits to calculate the monthly profit change                             
    for i in range(len(total_profit)-1):
        
        # The difference of the two months and tot profit to 
        monthly_profit_change.append(total_profit[i+1]-(total_profit[i]))
        

# The total number of months included in the dataset                          
print(f"Total months: {len(total_months)}")


# The net total amount of "Profit/Losses" over the entire period
print(f"Total amount: ${sum(total_profit)}")
           

#Calculation of the average of the changes in "Profit/Losses" over the entire period                                                     
average_change = sum(monthly_profit_change)/len(monthly_profit_change)
print(f"Average of the changes: {round(average_change,2)}")


# Max and Minof the monthly change list
increase_value = max(monthly_profit_change)
decrease_value = min(monthly_profit_change)

# The month associated with change is the next moth
greatest_increase = monthly_profit_change.index(min(monthly_profit_change))+1
greatest_decrease = monthly_profit_change.index(max(monthly_profit_change))+1

# FindThe greatest increase in profits (date and amount) over the entire period
print(f"Greatest increase:({total_months[greatest_increase]}) ${str(increase_value)}")

# The greatest decrease in losses (date and amount) over the entire period
print(f"Greatest decrease: ({total_months[greatest_decrease]}) ${str(decrease_value)}")


#Output to Text

with open("PyBank Final Results.txt", "w") as file:
    file.write(f"Total months: {len(total_months)}")
    file.write(f"\n")
    file.write(f"Total amount: ${sum(total_profit)}")
    file.write(f"\n")
    file.write(f"Average of the changes: {round(average_change,2)}")
    file.write(f"\n")
    file.write(f"Greatest increase:({total_months[greatest_increase]}) ${str(increase_value)}")
    file.write(f"\n")
    file.write(f"Greatest decrease: ({total_months[greatest_decrease]}) ${str(decrease_value)}")
               
    
```
